```python
### Task 1
from NNetwork import NNetwork as nn
G = nn.NNetwork()
G.load_add_edges("data/bn-mouse_retina_1.txt", use_genfromtxt=True, delimiter=",")
```

```python
### Learn a dictionary 
from ndl import NetDictLearner
NDL = NetDictLearner(G=G, n_components=30, k=25)
NDL.train_dict()
W = NDL.W
```

```python
print("Mouse: k=30, r=25.")
NDL.display_dict(path="output/mouse_retina_dic.jpg")
```

```python
### Learn fly's medulla's latent motifs
G_fly = nn.NNetwork()
G_fly.load_add_edges("data/bn-fly-drosophila_medulla_1.txt", 
                     use_genfromtxt=True, delimiter=",")
```

```python
from ndl import NetDictLearner
NDL_fly = NetDictLearner(G=G_fly, n_components=30, k=25)
NDL_fly.train_dict()
W_fly = NDL_fly.W
```

```python
print("Fly: k=30, r=25.")
NDL_fly.display_dict(path="output/fly_medulla_dic_1.jpg")
```

```python
### Task 2

NDL_fly_2 = NetDictLearner(G=G_fly, n_components=16, k=20)
NDL_fly_2.train_dict()
W_fly_2 = NDL_fly_2.W

NDL_fly_3 = NetDictLearner(G=G_fly, n_components=9, k=7)
NDL_fly_3.train_dict()
W_fly_3 = NDL_fly_3.W
```

```python
print("Fly: k=20, r=16.")
NDL_fly_2.display_dict(path="output/fly_medulla_dic_2.jpg")
```

```python
print("Fly: k=7, r=9.")
NDL_fly_3.display_dict(path="output/fly_medulla_dic_3.jpg")
```
